## Geometry

```python
import matplotlib.pyplot as plt

u = np.array([ -4, 4])

plt.figure(figsize = (6,6))

# 0, 0 intial point, u[0] e u[1] até onde vai
#
#

array1 = plt.arrow(0, 0, u[0], u[1], head_width = 0.2, width = 0.1, color = 'green',
length_includes_head = True)

#intial point
plt.plot(0, 0, 'ko', markerfacecolor = 'black', markersize = 6)

#plot 
plt.grid(linestyle = '--', linewidth = 0.5)
plt.axis('square')
plt.axis([-5,5,-5,5])
plt.legend([array1], ['u'])
plt.title('Array u')
plt.show()
```

## Array's operations

### Addition and Subtraction

```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
u = np.array([1, 2])
v = np.array([4, -6])
sum_vec = u + v
sub_vec = u - v

# Plotting the vectors
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot for addition
axs[0].arrow(0, 0, u[0], u[1], head_width=0.2, width=0.1, color='blue', length_includes_head=True)
axs[0].arrow(0, 0, v[0], v[1], head_width=0.2, width=0.1, color='black', length_includes_head=True)
axs[0].arrow(0, 0, sum_vec[0], sum_vec[1], head_width=0.2, width=0.1, color='pink', length_includes_head=True)
axs[0].plot(0, 0, 'ko', markerfacecolor='yellow', markersize=8)
axs[0].grid(linestyle='--', linewidth=0.6)
axs[0].axis('square')
axs[0].axis([-10, 10, -10, 10])
axs[0].legend(['u', 'v', 'u + v'], loc='upper left')
axs[0].set_title('Vectors $\mathbf{u}$, $\mathbf{v}$, and $\mathbf{u + v}$')

# Plot for subtraction
axs[1].arrow(0, 0, u[0], u[1], head_width=0.2, width=0.1, color='blue', length_includes_head=True)
axs[1].arrow(0, 0, v[0], v[1], head_width=0.2, width=0.1, color='black', length_includes_head=True)
axs[1].arrow(0, 0, sub_vec[0], sub_vec[1], head_width=0.2, width=0.1, color='green', length_includes_head=True)
axs[1].plot(0, 0, 'ko', markerfacecolor='yellow', markersize=8)
axs[1].grid(linestyle='--', linewidth=0.6)
axs[1].axis('square')
axs[1].axis([-10, 10, -10, 10])
axs[1].legend(['u', 'v', 'u - v'], loc='upper left')
axs[1].set_title('Vectors $\mathbf{u}$, $\mathbf{v}$, and $\mathbf{u - v}$')

# Show the plot
plt.tight_layout()
plt.show()

```

### Multiplying a vector by a scalar

```python

# List of scalar
scalar = [ 1, 2, 1/2, -1, -2 ]

# Vetor
initial_vector = np.array([ .9, 1.2 ])


# Figure
fig, axs = plt.subplots(1, len(scalar), figsize = (14,5))

# Contador para os subplots
i = 0 

for valor in scalar:
    
    # scalar
    resultant_vector = valor * initial_vector

    # Plot do vetor inicial
    axs[i].arrow(0, 
                 0, 
                 initial_vector[0], 
                 initial_vector[1], 
                 head_width = .3,
                 width = .1, 
                 color = 'blue',
                 length_includes_head = True)
    
    # resultant vector
    axs[i].arrow(0.1, 
                 0, 
                 resultant_vector[0], 
                 resultant_vector[1], 
                 head_width = .3,
                 width = .1,
                 color = 'red',
                 length_includes_head = True)
    
    
    axs[i].grid(linestyle = '--')
    axs[i].axis('square')
    axs[i].axis([-2.5, 2.5, -2.5, 2.5])
    
    # Lim
    axs[i].set(xticks = np.arange(-2,3), yticks = np.arange(-2,3))
    
    # Title
    axs[i].set_title(f'$\sigma$ = {valor:.2f}')
    
    # Contador
    i += 1

plt.show()
```

### Dot product

#### Super useful - how similar two normailized vector they are? 

```python
A = np.array([ 5, 2, 1])
B = np.array([ 3, 2, 4 ])
```

```python
# normalized

A_normalized = A / np.linalg.norm(A)
B_normalized = B / np.linalg.norm(B)
```

```python
result1 = np.dot(A_normalized, B_normalized)
print(result1)
```

```python
# three vectors

u = np.array([1, 5])
v = np.array([6, 2])
w = np.array([9, 4])
```

```python
result_1 = np.dot(u, v+w)
print(result_1)
```

```python
result2 = np.dot(u, v) + np.dot(u, w)
print(result2)
```

```python

vw_sum = v + w


plt.figure()


plt.quiver(0, 0, u[0], u[1], color = 'r', scale = 1, scale_units = 'xy', angles = 'xy', label = 'u')
plt.quiver(0, 0, v[0], v[1], color = 'b', scale = 1, scale_units = 'xy', angles = 'xy', label = 'v')
plt.quiver(0, 0, w[0], w[1], color = 'g', scale = 1, scale_units = 'xy', angles = 'xy', label = 'w')
plt.quiver(0, 0, vw_sum[0], vw_sum[1], color = 'orange', scale = 1, scale_units = 'xy', angles = 'xy', label = 'v+w')


plt.xlim(-10, 20)
plt.ylim(-10, 20)
plt.grid()
plt.legend()
plt.show()
```

### Cross product

```python
result_cross = np.cross(A, B)
print(result_cross)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Define the vectors
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])


result_cross = np.cross(A, B)

# Plotting in 3D
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot vectors
ax.quiver(0, 0, 0, A[0], A[1], A[2], color='r', label='A')
ax.quiver(0, 0, 0, B[0], B[1], B[2], color='g', label='B')
ax.quiver(0, 0, 0, result_cross[0], result_cross[1], result_cross[2], color='b', label='Cross Product (A × B)')


ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([-10, 10])
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Visualization of Cross Product')
ax.legend()

plt.show()
```
