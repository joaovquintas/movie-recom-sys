## Geometry

```python
import matplotlib.pyplot as plt

u = np.array([ -4, 4])

plt.figure(figsize = (6,6))

# 0, 0 intial point, u[0] e u[1] at√© onde vai
#
#

array1 = plt.arrow(0, 0, u[0], u[1], head_width = 0.2, width = 0.1, color = 'green',
length_includes_head = True)

#intial point
plt.plot(0, 0, 'ko', markerfacecolor = 'black', markersize = 6)

#plot 
plt.grid(linestyle = '--', linewidth = 0.5)
plt.axis('square')
plt.axis([-5,5,-5,5])
plt.legend([array1], ['u'])
plt.title('Array u')
plt.show()
```

```python
# Exemple 2 

plt.figure(figsize = (6,6))


arrays = np.array([[2, -3],[4, -1], [-4, 2]])

#origin
origins = np.array([[0, 1, -1],[-1, -2, 1]])

#plot - quiver

plt.quiver(*origins, arrays[:,0], arrays[:,1], color = ['r', 'g', 'b'], scale = 20)

plt.axis([-5,5,-5,5])
plt.show()
```

## Array's operations

### Addition and Subtraction

```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
u = np.array([1, 2])
v = np.array([4, -6])
sum_vec = u + v
sub_vec = u - v

# Plotting the vectors
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot for addition
axs[0].arrow(0, 0, u[0], u[1], head_width=0.2, width=0.1, color='blue', length_includes_head=True)
axs[0].arrow(0, 0, v[0], v[1], head_width=0.2, width=0.1, color='black', length_includes_head=True)
axs[0].arrow(0, 0, sum_vec[0], sum_vec[1], head_width=0.2, width=0.1, color='pink', length_includes_head=True)
axs[0].plot(0, 0, 'ko', markerfacecolor='yellow', markersize=8)
axs[0].grid(linestyle='--', linewidth=0.6)
axs[0].axis('square')
axs[0].axis([-10, 10, -10, 10])
axs[0].legend(['u', 'v', 'u + v'], loc='upper left')
axs[0].set_title('Vectors $\mathbf{u}$, $\mathbf{v}$, and $\mathbf{u + v}$')

# Plot for subtraction
axs[1].arrow(0, 0, u[0], u[1], head_width=0.2, width=0.1, color='blue', length_includes_head=True)
axs[1].arrow(0, 0, v[0], v[1], head_width=0.2, width=0.1, color='black', length_includes_head=True)
axs[1].arrow(0, 0, sub_vec[0], sub_vec[1], head_width=0.2, width=0.1, color='green', length_includes_head=True)
axs[1].plot(0, 0, 'ko', markerfacecolor='yellow', markersize=8)
axs[1].grid(linestyle='--', linewidth=0.6)
axs[1].axis('square')
axs[1].axis([-10, 10, -10, 10])
axs[1].legend(['u', 'v', 'u - v'], loc='upper left')
axs[1].set_title('Vectors $\mathbf{u}$, $\mathbf{v}$, and $\mathbf{u - v}$')

# Show the plot
plt.tight_layout()
plt.show()

```

### Multiplying a vector by a scalar

```python
# Efeito de diferentes valores escalares

# Lista de escalaraes
valores_escalares = [ 1, 2, 1/2, -1, -2 ]

# Vetor
vetor_inicial = np.array([ .9, 1.2 ])


# Cria a figura
fig, axs = plt.subplots(1, len(valores_escalares), figsize = (14,5))

# Contador para os subplots
i = 0 

for valor in valores_escalares:
    
    # scalar
    vetor_resultante = valor * vetor_inicial

    # Plot do vetor inicial
    axs[i].arrow(0, 
                 0, 
                 vetor_inicial[0], 
                 vetor_inicial[1], 
                 head_width = .3,
                 width = .1, 
                 color = 'blue',
                 length_includes_head = True)
    
    # resultant vector
    axs[i].arrow(0.1, 
                 0, 
                 vetor_resultante[0], 
                 vetor_resultante[1], 
                 head_width = .3,
                 width = .1,
                 color = 'red',
                 length_includes_head = True)
    
    # Grid
    axs[i].grid(linestyle = '--')
    
    # Formato
    axs[i].axis('square')
    
    # Eixos
    axs[i].axis([-2.5, 2.5, -2.5, 2.5])
    
    # Limites
    axs[i].set(xticks = np.arange(-2,3), yticks = np.arange(-2,3))
    
    # T√≠tulo
    axs[i].set_title(f'$\sigma$ = {valor:.2f}')
    
    # Contador
    i += 1

plt.show()
```
